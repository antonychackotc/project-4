```python
%%writefile app1.py

import streamlit as st
import numpy as np
import pandas as pd
import joblib

# Load saved models and encoders
log_transformer_total = joblib.load("log_transformer_TotalWorkingYears.pkl")
log_transformer_year = joblib.load("log_transformer_YearsAtCompany.pkl")
log_transformer_promo = joblib.load("log_transformer_YearsSinceLastPromotion.pkl")
label_encoders = joblib.load("label_encoders.pkl")
freq_encoding_map_job = joblib.load("jobrole_freq_encoding.pkl")
ohe_columns = joblib.load("ohe_columns.pkl")

# Load trained Random Forest model
loaded_model = joblib.load("random_forest_regressor_model.pkl")

# Function to categorize tenure
def tenure_category(years):
    if years < 2:
        return 'New'
    elif years < 5:
        return 'Intermediate'
    elif years < 10:
        return 'Experienced'
    else:
        return 'Veteran'

# Streamlit UI
st.title("Employee Monthly Income Prediction")
st.write("Enter employee details to predict their monthly income.")

# User Inputs
job_level = st.number_input("Job Level", min_value=1, max_value=5, step=1)
total_working_years = st.number_input("Total Working Years", min_value=0, step=1)
age = st.number_input("Age", min_value=18, max_value=65, step=1)
years_at_company = st.number_input("Years at Company", min_value=0, step=1)
job_role = st.selectbox("Job Role", list(freq_encoding_map_job.keys()))
years_in_role = st.number_input("Years in Current Role", min_value=0, step=1)
years_with_manager = st.number_input("Years with Current Manager", min_value=0, step=1)
years_since_promotion = st.number_input("Years Since Last Promotion", min_value=0, step=1)

# Preprocessing
new_data = pd.DataFrame({
    'JobLevel': [job_level],
    'TotalWorkingYears': [total_working_years],
    'Age': [age],
    'YearsAtCompany': [years_at_company],
    'JobRole_FreqEncoded': [freq_encoding_map_job.get(job_role, 0)],
    'YearsInCurrentRole': [years_in_role],
    'YearsWithCurrManager': [years_with_manager],
    'YearsSinceLastPromotion': [years_since_promotion]
})

# Apply log transformations
new_data['TotalWorkingYears'] = log_transformer_total.transform(new_data[['TotalWorkingYears']])
new_data['YearsAtCompany'] = log_transformer_year.transform(new_data[['YearsAtCompany']])
new_data['YearsSinceLastPromotion'] = log_transformer_promo.transform(new_data[['YearsSinceLastPromotion']])

# Tenure Category
new_data['Tenure_Category'] = new_data['YearsAtCompany'].apply(tenure_category)
if 'Tenure_Category' in label_encoders:
    new_data['Tenure_Category'] = label_encoders['Tenure_Category'].transform(new_data['Tenure_Category'])
else:
    new_data = pd.get_dummies(new_data, columns=['Tenure_Category'], drop_first=True)

# Ensure all expected columns exist
for col in ohe_columns:
    if col not in new_data:
        new_data[col] = 0  # Add missing columns

# Reorder to match training data
required_features = ['JobLevel', 'TotalWorkingYears', 'Age', 'YearsAtCompany',
                     'JobRole_FreqEncoded', 'YearsInCurrentRole', 'Tenure_Category',
                     'YearsWithCurrManager', 'YearsSinceLastPromotion']

new_data = new_data.reindex(columns=required_features, fill_value=0)

# Prediction
if st.button("Predict Monthly Income"):
    prediction = loaded_model.predict(new_data)
    st.success(f"Predicted Monthly Income: ${prediction[0]:,.2f}")
```
